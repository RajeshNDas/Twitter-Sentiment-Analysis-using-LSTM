```python
df=df.drop(columns=['Unnamed: 1','Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
```

```python
df.info()
```

```python
df['sentiment'] = df['sentiment'].map({0:0, 4:1})
```

```python
df
```

```python
df['text']=df['text'].str.lower()
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
swords=stopwords.words('english')
```

```python
swords
```

```python
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in swords])

df['text'] = df['text'].apply(lambda text: remove_stopwords(text))
df['text'].head(20)
```

```python
import string

def remove_punctuations(text):
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)

df['text']= df['text'].apply(lambda x: remove_punctuations(x))
df['text'].head(20)
```

```python
import re
def remove_URLs(data):
    return re.sub('((www.[^s]+)|(https?://[^s]+))',' ',data)

df['text'] = df['text'].apply(lambda x: remove_URLs(x))
df['text'].tail()
```

```python
def remove_numbers(data):
    return re.sub('[0-9]+', '', data)

df['text'] = df['text'].apply(lambda x: remove_numbers(x))
df['text'].head(20)
```

```python
sentences = df['text'].tolist()
labels = df['sentiment'].tolist()
```

```python
sentences
```

```python
print(sentences[0])
print(labels[0])
```

```python
#Tokenize the sentences
tokenizer = Tokenizer()

#preparing vocabulary
tokenizer.fit_on_texts(sentences)
word_index = tokenizer.word_index

size_of_vocabulary=len(word_index) + 1 
print("Vocabulary Size = %d"%size_of_vocabulary)
```

```python
# split the data into a training set and a validation set
indices = np.arange(len(sentences))
print(indices)
#print(len(indices))
np.random.shuffle(indices)
print(indices)
#print(len(indices))
```

```python
sentences=np.array(sentences)[indices.astype(int)]
labels=np.array(labels)[indices.astype(int)]
```

```python
sentences[500]
```

```python
labels[500]
```

```python
training_size = int(len(sentences) * 0.9)

training_sentences = sentences[0: training_size]
testing_sentences = sentences[training_size:]
training_labels = labels[0: training_size]
testing_labels = labels[training_size:]

# Put labels into list to use later:

training_labels_final = np.array(training_labels)
testing_labels_final = np.array(testing_labels)
```

```python
print("Shape of training dataset X : "+str(len(training_sentences)))
print("Shape of training dataset Y : "+str(len(training_labels_final)))
print("Shape of testing dataset X : "+str(len(testing_sentences)))
print("Shape of testing dataset Y : "+str(len(testing_labels_final)))
```

```python
embedding_dim = 100
max_length = 100
trunc_type='post'
padding_type='post'
```

```python
training_sequences = tokenizer.texts_to_sequences(training_sentences)
training_padded = pad_sequences(training_sequences, maxlen=max_length, padding=padding_type, truncating=trunc_type)

testing_sequences = tokenizer.texts_to_sequences(testing_sentences)
testing_padded = pad_sequences(testing_sequences, maxlen=max_length, padding=padding_type, truncating=trunc_type)
```

```python
print(training_sentences[20])
print(training_sequences[20])
training_padded[20]
```

```python
embeddings_index = dict()
f = open('/content/drive/MyDrive/glove.twitter.27B.100d.txt')

for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = coefs

f.close()
print('Loaded %s word vectors.' % len(embeddings_index))
```

```python
# create a weight matrix for words in training docs
embedding_matrix = np.zeros((size_of_vocabulary, 100))

for word, i in tokenizer.word_index.items():
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector
```

```python
embedding_matrix.shape
```

```python
from keras.models import *
from keras.layers import *
from keras.callbacks import *

embedding_layer = Embedding(size_of_vocabulary, embedding_dim, weights = [embedding_matrix], input_length = max_length, trainable = False)
model = Sequential()
model.add(embedding_layer)

model.add(LSTM(200, dropout = 0.2))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation = "sigmoid"))

#Add loss function, metrics, optimizer
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=["acc"]) 

#Adding callbacks
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1,patience=3)  
mc=ModelCheckpoint('best_model.h5', monitor='val_acc', mode='max', save_best_only=True,verbose=1)
reduce_lr = ReduceLROnPlateau(monitor = 'val_acc', factor = 0.1,min_lr = 0.01,verbose=1)
```

```python
#Print summary of model
print(model.summary())
```

```python
history = model.fit(training_padded,training_labels_final,batch_size=1024,epochs=15,validation_data=(testing_padded,testing_labels_final),verbose=1,callbacks=[es,mc,reduce_lr])
```

```python
with open('model_architecture_Twitter_classifier_second_try.json','w') as f:
    f.write(model.to_json())
```
